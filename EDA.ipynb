```python
min_odds_lst = [col for col in odds.columns[2:] if 'Min' in col]
min_odds_lst  
```

```python
max_odds_lst = [col for col in odds.columns[2:] if 'Max' in col]
max_odds_lst
```

#### Make Range of Odds Columns

```python
def range_of_odds(df, col_1,col_2, type_of_odds):
    df[type_of_odds+' Range'] = abs(df[col_1]-df[col_2])
```

```python
min_max_odds_lst = list(zip(min_odds_lst, max_odds_lst))
col_names = ['Home Odds', 'Away Odds', 'Home Line Odds', 'Away Line Odds', 'Total Score Over Odds', 'Total Score Under Odds']
for i in range(len(min_max_odds_lst)):
    range_of_odds(odds, min_max_odds_lst[i][0],min_max_odds_lst[i][1], col_names[i])
    
```
